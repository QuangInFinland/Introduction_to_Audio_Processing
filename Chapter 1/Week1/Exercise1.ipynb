```python
# Problem 1: Create a synthetic signal as a sum of sinusoids. (1 point)

# a) Create 4 sinusoids of 3 s length, with different amplitude, frequencies 100, 500, 1500,
# 2500 Hz, different phases, sampled at 8kHz. (Hint: how many samples the signal must
# have for 3 s length? take into account fs).
# b) Play and plot the sinusoids
# c) Add them up to x(t). Plot and play x(t). Write the signal to a wav file.
# d) Apply DFT (you can use scipy.fftpack.fft function. Select the number of DFT points,
# e.g., to 512)). Plot magnitude DFT.
# e ) Observe the components and relationship between nfft and frequency in Hz. (DFT has
# imaginary values x+iy, where x is amplitude information and y is phase). Report your
# observation.

import numpy as np
import matplotlib.pyplot as plot
import scipy.io.wavfile as wavfile
import scipy.fftpack as fftpack
import scipy
import sounddevice as sd


fs= 8000
f1 = 100
f2 = 500
f3 = 1500
f4 = 2500




plot.subplot(4,1,1)
plot.title('Sinusoidal signal with frequency 100Hz')
A = 5
t = np.linspace(0,3,fs*3)
ph1 = np.pi/2
y1 = A*np.sin(2*np.pi*f1*t + ph1)
plot.plot(t[0:300],y1[0:300])
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.show()
sd.play(y1,fs)
sd.wait()

plot.subplot(4,1,2)
plot.title('Sinusoidal signal with frequency 500Hz')
A = 5
t = np.linspace(0,3,fs*3)
ph2 = np.pi
y2 = A*np.sin(2*np.pi*f2*t + ph2)
plot.plot(t[0:300],y2[0:300])
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.show()
sd.play(y2,fs)
sd.wait()


plot.subplot(4,1,3)
plot.title('Sinusoidal signal with frequency 1000Hz')
A = 5
t = np.linspace(0,3,fs*3)
ph3 = 3*np.pi/2
y3 = A*np.sin(2*np.pi*f3*t + ph3)
plot.plot(t[0:300],y3[0:300])
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.show()
sd.play(y3,fs)
sd.wait()

plot.subplot(4,1,4)
plot.title('Sinusoidal signal with frequency 2500Hz')
A = 5
t = np.linspace(0,3,fs*3)
ph4 = 2*np.pi
y4 = A*np.sin(2*np.pi*f4*t + ph4)
plot.plot(t[0:300],y4[0:300])
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.show() 
sd.play(y4,fs)
sd.wait()

plot.title('Sum of all the signals')
x = y1+y2+y3+y4
t = np.linspace(0,3,fs*3)
plot.plot(t[0:300],x[0:300])
plot.xlabel('Time')
plot.ylabel('Amplitude')
plot.show()
sd.play(x,fs)
sd.wait()

wavfile.write('sum.wav',fs,x.astype(np.int16))


# Applying DFT
N = 512
X = fftpack.fft(x,N)
x = np.arange(N)

# Plotting the magnitude spectrum
plot.figure()
magnitude = np.abs(X)
plot.plot(x, magnitude)
plot.title('DFT Magnitude')
plot.xlabel('Frequency (Hz)')
plot.ylabel('Magnitude')
plot.show()

print(X)

# Observation: 512 samples are on the plot. the first and last points values are the same lowest amplitude which over 0 a little bit.
# The second and second last points are the highest amplitude. The third and third last points are the second highest amplitude.











```
